# Creating Barcharts using matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

col_count = 5
bar_width = 0.15

aspop = [60.9, 60.8, 60.5, 60.1, 59.5]
afpop = [13.2, 14.0, 14.9, 16.0, 17.2]
eupop = [11.8, 11.2, 10.6, 10.1, 9.6]
lapop = [8.5, 8.5, 8.5, 8.5, 8.4]
napop = [5.1, 5.0, 4.9, 4.8, 4.7]

index = np.arange(col_count)

as1 = plt.bar(index, aspop, bar_width, alpha=0.4, label='Asia')
af1 = plt.bar(index + 0.15, afpop, bar_width, alpha=0.4, label='Africa')
eu1 = plt.bar(index + 0.30, eupop, bar_width, alpha=0.4, label='Europe')
la1 = plt.bar(index + 0.45, lapop, bar_width, alpha=0.4, label='Latin America')
na1 = plt.bar(index + 0.60, napop, bar_width, alpha=0.4, label='Noth America')



def CreateLabels(data):
    for item in data:
        height = item.get_height()
        plt.text(item.get_x() + item.get_width()/2., height*1.05,
                 '%d' % int(height), ha='center', va='bottom')

CreateLabels(as1)
CreateLabels(af1)
CreateLabels(eu1)
CreateLabels(la1)
CreateLabels(na1)



plt.xticks(index + 0.6/2, ("2000", "2005", "2010", "2015", "2020"))

plt.ylabel("Population Growth in %")
plt.xlabel("Year >>>")


plt.title("Population of the World")
plt.legend(frameon=True, loc=1)

plt.show()
```
